```python
#Extraer la información del archivo. 
import pandas as pd
df = pd.read_csv('/country_vaccinations.csv', sep=None)
print (df)

```

```python
df[['date']]= df[['date']].apply(pd.to_datetime)
```

```python
# Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64. 
df.info()
```

```python
#Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines). 
df.groupby(['vaccines']).count()
```

```python
#Obtener la cantidad de vacunas aplicadas en todo el mundo. 
Totalvacunados = df['total_vaccinations'].sum()
print(Totalvacunados)
```

```python
# Calcular el promedio de vacunas aplicadas por país. 

df.groupby(['country','vaccines']).mean()

```

```python
# Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.
df[df["date"] == "29/01/21"].sum()
```

```python
#Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw
Diferencias= df['daily_vaccinations'] - df['daily_vaccinations_raw']
print(Diferencias)
```

```python
#Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  
import pandas as pd

conDiferencias = pd.concat([df['daily_vaccinations'],df['daily_vaccinations_raw'], Diferencias], axis=1)
pd.DataFrame(conDiferencias)




```

```python
#h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua. 
fechareciente = df['date'].max()
fechalejana = df['date'].min()

print(fechareciente)
print(fechalejana)

periodo = fechareciente - fechalejana
print(periodo)
```

```python
#i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter /). 

canVan = df["vaccines"].str.split('/', expand= True)
print(canVan)
import pandas as pd
conCantidad = pd.concat([df, canVan], axis = 1)
pd.DataFrame(conCantidad)
```

```python
#j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020. 
filtro = (df['date'] > '2020-12-20')
antes20 = df.loc[filtro]
pd.DataFrame(antes20)
```

```python
from pandas.core.frame import DataFrame
#k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer. 
df['vaccines']
print(df['vaccines'])

pfizer=  df[df.set_index('vaccines').filter(regex= 'Pzfizer', axis=1)]
pd.DataFrame(pfizer)


```

```python
#l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter. 
import pandas as pd
import numpy as np


df1 = pd.DataFrame(conDiferencias)  
df2 = pd.DataFrame(conCantidad)
df3 = pd.DataFrame(antes20)
df4 = pd.DataFrame(pfizer)
with pd.ExcelWriter("resultadosReto.xlsx") as writer:
      df1.to_excel(writer, sheet_name="conDiferencias")  
      df2.to_excel(writer, sheet_name="conCantidad")
      df3.to_excel(writer, sheet_name="antes20")
      df4.to_excel(writer, sheet_name="pfizer")
print('Please check out subject-wise resultadosReto.xlsx file.')
```
